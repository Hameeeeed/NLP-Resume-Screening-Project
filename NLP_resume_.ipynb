```python
import nltk
import string
import pandas as pd
import PyPDF2

from google.colab import files
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
def preprocess_text(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    return text
```

```python
def extract_text_from_pdf(pdf_file):
    text = ""
    reader = PyPDF2.PdfReader(pdf_file)
    for page in reader.pages:
        if page.extract_text():
            text += page.extract_text()
    return text
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
job_description = """
We are looking for a data analyst with experience in machine learning,
data analysis, NLP, Pandas, NumPy, Opencv,DLIb and deep learning.
"""

job_description = preprocess_text(job_description)
job_embedding = model.encode(job_description)
```

```python
print("Upload one or more resume PDFs")
uploaded = files.upload()

```

```python
results = []

for filename in uploaded.keys():
    resume_text = extract_text_from_pdf(filename)
    resume_text = preprocess_text(resume_text)

    resume_embedding = model.encode(resume_text)

    similarity_score = cosine_similarity(
        [job_embedding], [resume_embedding]
    )[0][0]

    match_percentage = similarity_score * 100
    status = "Shortlisted ✅" if similarity_score >= 0.65 else "Not Shortlisted ❌"

    results.append({
        "Resume Name": filename,
        "Match Percentage": round(match_percentage, 2),
        "Status": status
    })
```

```python
df = pd.DataFrame(results)
df.sort_values(by="Match Percentage", ascending=False)
```
